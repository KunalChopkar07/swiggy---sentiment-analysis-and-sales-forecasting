CITIES NAME

```python
most_popular_cuisines = {}
for city in cities:
    city_df = df[df['City'] == city]
    cuisine_counts = city_df['Food type'].value_counts()
    most_popular_cuisine = cuisine_counts.index[0]
    most_popular_cuisines[city] = most_popular_cuisine
for city, most_popular_cuisine in most_popular_cuisines.items():
    print('The most popular cuisine in {} is {}'.format(city, most_popular_cuisine))
```

CALCULATINF THE MOST FAMOUS FOOD TYPES IN THE CITY

```python
restaurant_types = df['Food type'].unique()
restaurant_types_and_prices_df = pd.DataFrame({'Restaurant type': restaurant_types,'Average meal price': df.groupby('Food type')['Price'].mean()})
restaurant_types_and_prices_df = restaurant_types_and_prices_df.sort_values(by='Average meal price', ascending=False)
restaurant_types_and_prices_df = restaurant_types_and_prices_df.reset_index()
print(restaurant_types_and_prices_df.head(5).to_string(index=True))
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Load your CSV
df = pd.read_csv('swiggy_final_data.csv')

# Create Quantity column (assuming 1 or random between 1-3)
df['Quantity'] = [random.randint(1, 3) for _ in range(len(df))]

# Create Order_ID column
df['Order_ID'] = ['ORD' + str(1000 + i) for i in range(len(df))]

# Create Date column (random dates between Jan 1, 2024 and Apr 30, 2024)
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 4, 30)
df['Date'] = [start_date + timedelta(days=random.randint(0, (end_date - start_date).days)) for _ in range(len(df))]

# If you don't have Sentiment, you can randomly assign
sentiments = ['Positive', 'Negative', 'Neutral']
df['Sentiment'] = [random.choice(sentiments) for _ in range(len(df))]

# Save final file
df.to_csv('swiggy_final_data.csv', index=False)

print("âœ… Final CSV with all required columns created!")
```
